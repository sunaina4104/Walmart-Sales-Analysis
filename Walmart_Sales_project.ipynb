```python
# importing the libraries pandas,matplotlib and seaborn 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import Dataset of Walmart Sales
df = pd.read_csv(r"C:\Users\user\Downloads\Walmart_Sales.csv")
```

```python
#  total weekly sales of Walmart 
total_sales = df["Weekly_Sales"].sum()
print(total_sales)
```

```python
# average weekly sales of Walmart 
average_sales = df["Weekly_Sales"].mean()
print(average_sales)
```

```python
# total sales by Store number 
store_sales = df.groupby("Store")["Weekly_Sales"].sum()
print(store_sales)
```

```python
df1 = df.sort_values(by="Date")
df1
```

```python
# to check the total sales according to holiday and non holiday weeks
holiday_wise_sales_total = df.groupby("Holiday_Flag")["Weekly_Sales"].sum()
print(holiday_wise_sales_total)
```

```python
# to check the average sales according to holiday and non holiday weeks
holiday_wise_sales_avg = df.groupby("Holiday_Flag")["Weekly_Sales"].mean()
print(holiday_wise_sales_avg)
```

```python
# to convert the date format in actual time & date format that python understands  
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')
```

```python
# create line chart to check weekly sales on the basis of number of stores 
plt.figure()
plt.plot(df["Store"],df["Weekly_Sales"])
plt.xlabel("Store")
plt.ylabel("Weekly Sales")
plt.title("Weekly sales on each store")
plt.show()
```

```python
plt.figure()
plt.bar(df["Store"],df["Weekly_Sales"],color="navy")
plt.xlabel("Store")
plt.ylabel("Weekly Sales")
plt.title("Sales trend")
plt.show()
```

```python
# 
store_wise_salesavg = df.groupby('Store')['Weekly_Sales'].mean()
store_wise_salesavg.plot(kind='bar',color = 'pink')
plt.xlabel("Store")
plt.ylabel("Average Weekly Sales")
plt.title("Store wise average sales")
plt.show()
```

```python
sns.scatterplot(x='Temperature', y='Weekly_Sales', data=df, color = 'red')
plt.title("Temperature vs Weekly Sales")
plt.show()
```

```python
# 
plt.figure()
plt.hist(df["Weekly_Sales"],color = 'brown')
plt.xlabel("Weekly Sales")
plt.title("Weekly sales Sales trend")
plt.show()
```

```python
store_sales = df.groupby("Store")["Weekly_Sales"].sum()

top10_stores = store_sales.sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
plt.bar(top10_stores.index, top10_stores.values, color="pink")

plt.xlabel("Store")
plt.ylabel("Total Weekly Sales")
plt.title("Top 10 Stores by Total Sales")
plt.show()
```

```python
holiday_wise_salesavg = df.groupby('Holiday_Flag')['Weekly_Sales'].mean()

holiday_wise_salesavg.plot(kind='bar' , color =['navy', 'pink'] )

plt.xlabel("Holiday Flag (0 = Non-Holiday, 1 = Holiday)")
plt.ylabel("Average Weekly Sales")
plt.title("Holiday vs Non-Holiday Average Sales")
plt.show()
```

```python
plt.figure()
sns.boxplot(y=df['Weekly_Sales'],color = 'yellow')
plt.title("Overall Weekly Sales Distribution")
plt.ylabel("Weekly Sales")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numeric columns
corr = df[['Temperature', 'Weekly_Sales']].corr()

# Create heatmap
plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True, cmap='Blues', fmt='.2f')

plt.title("Correlation Heatmap: Temperature vs Weekly Sales")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Holiday_Flag', data=df, color = 'orange')

plt.xlabel("Holiday Flag (0 = Non-Holiday, 1 = Holiday)")
plt.ylabel("Count")
plt.title("Count of Holiday vs Non-Holiday Weeks")
plt.show()
```

```python
plt.hexbin(df['Temperature'], df['Weekly_Sales'], gridsize=30,color = 'pink')
plt.colorbar(label='frequency')
plt.xlabel("Temperature")
plt.ylabel("Weekly Sales")
plt.title("Density of Sales Across Temperature")
plt.show()
```
